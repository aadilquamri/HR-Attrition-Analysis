```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Define a color sequence for consistency
color_sequence = px.colors.qualitative.Set2

# Plot the distribution of numerical columns using colorful histograms
numerical_columns = ['Age', 'JobSatisfaction', 'MonthlyIncome', 'PercentSalaryHike', 
                     'TotalWorkingYears', 'YearsAtCompany', 'YearsSinceLastPromotion']

for column in numerical_columns:
    fig = px.histogram(
        df, x=column, nbins=20,
        title=f'Distribution of {column}',
        color_discrete_sequence=color_sequence
    )
    fig.update_layout(
        xaxis_title=column,
        yaxis_title='Count',
        template='plotly',
        bargap=0.1
    )
    fig.show()

# Boxplot for numerical columns vs Attrition
for col in ['MonthlyIncome', 'Age']:
    fig = px.box(
        df, x='Attrition', y=col, color='Attrition',
        title=f'Attrition vs {col}',
        color_discrete_sequence=color_sequence
    )
    fig.update_layout(
        xaxis_title='Attrition',
        yaxis_title=col,
        template='plotly'
    )
    fig.show()

# Scatter plots with better visuals
fig = px.scatter(
    df, x='Age', y='MonthlyIncome', color='Attrition',
    title="Age vs Monthly Income (by Attrition)",
    color_discrete_sequence=color_sequence,
    marginal_x='histogram', marginal_y='violin', opacity=0.7
)
fig.update_layout(
    xaxis_title='Age',
    yaxis_title='Monthly Income',
    template='plotly'
)
fig.show()

fig = px.scatter(
    df, x='YearsAtCompany', y='TotalWorkingYears', color='Attrition',
    title="Years at Company vs Total Working Years (by Attrition)",
    color_discrete_sequence=color_sequence,
    marginal_x='box', marginal_y='rug', opacity=0.7
)
fig.update_layout(
    xaxis_title='Years At Company',
    yaxis_title='Total Working Years',
    template='plotly'
)
fig.show()
```

```python
import plotly.express as px
import plotly.subplots as sp
import plotly.graph_objects as go

fig = sp.make_subplots(rows=1, cols=3, subplot_titles=[
    "Age vs MonthlyIncome", "Age vs DistanceFromHome", "MonthlyIncome vs DistanceFromHome"
])

fig.add_trace(go.Scatter(
    x=df["Age"], y=df["MonthlyIncome"],
    mode="markers", marker=dict(color=df['Attrition'].astype('category').cat.codes),
    name="Age vs Income"
), row=1, col=1)

fig.add_trace(go.Scatter(
    x=df["Age"], y=df["DistanceFromHome"],
    mode="markers", marker=dict(color=df['Attrition'].astype('category').cat.codes),
    name="Age vs Distance"
), row=1, col=2)

fig.add_trace(go.Scatter(
    x=df["MonthlyIncome"], y=df["DistanceFromHome"],
    mode="markers", marker=dict(color=df['Attrition'].astype('category').cat.codes),
    name="Income vs Distance"
), row=1, col=3)

# Axis labels for 1st subplot (normal)
fig.update_xaxes(title_text="Age", row=1, col=1)
fig.update_yaxes(title_text="MonthlyIncome", row=1, col=1)

# Axis labels for 2nd subplot (normal)
fig.update_xaxes(title_text="Age", row=1, col=2)
fig.update_yaxes(title_text="DistanceFromHome", row=1, col=2)

# Axis labels for 3rd subplot: y-axis label on right side
fig.update_xaxes(title_text="MonthlyIncome", row=1, col=3)
fig.update_yaxes(title_text="DistanceFromHome", row=1, col=3, side='right')

fig.update_layout(title="Scatter Plots by Attrition", showlegend=False)
fig.show()
fig.write_image("plot_2.pdf", width=1000, height=400, scale=8)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import LabelEncoder
import numpy as np

# --- Dummy DataFrame Creation (REPLACE THIS WITH YOUR ACTUAL DATAFRAME LOADING) ---
# This section is for demonstration purposes only, mimicking your df_1 structure.
# In your actual code, you would load your DataFrame like:
# df = pd.read_csv("your_hr_data.csv")
# Or ensure your existing 'df' has the necessary columns.

# Creating a dummy DataFrame that closely resembles your df_1.columns structure
# Note: One-hot encoded columns are included for completeness, but not directly used
# in the current visualizations as 'JobRole' and 'Department' are still present.
data = {
    'Age': np.random.randint(20, 60, 500),
    'Attrition': np.random.choice(['Yes', 'No'], 500, p=[0.15, 0.85]),
    'Gender': np.random.choice(['Male', 'Female'], 500),
    'JobSatisfaction': np.random.randint(1, 5, 500), # 1: Low, 2: Medium, 3: High, 4: Very High
    'MonthlyIncome': np.random.randint(2000, 15000, 500),
    'OverTime': np.random.choice(['Yes', 'No'], 500, p=[0.3, 0.7]), # Correct casing 'OverTime'
    'PercentSalaryHike': np.random.randint(10, 25, 500),
    'TotalWorkingYears': np.random.randint(0, 30, 500),
    'WorkLifeBalance': np.random.randint(1, 5, 500),
    'YearsAtCompany': np.random.randint(0, 20, 500),
    'YearsSinceLastPromotion': np.random.randint(0, 15, 500),
    'BusinessTravel_Travel_Frequently': np.random.randint(0, 2, 500), # Dummy one-hot
    'BusinessTravel_Travel_Rarely': np.random.randint(0, 2, 500),     # Dummy one-hot
    'Department_Research & Development': np.random.randint(0, 2, 500), # Dummy one-hot
    'Department_Sales': np.random.randint(0, 2, 500),                 # Dummy one-hot
    'JobRole_Human Resources': np.random.randint(0, 2, 500),          # Dummy one-hot
    'JobRole_Laboratory Technician': np.random.randint(0, 2, 500),
    'JobRole_Manager': np.random.randint(0, 2, 500),
    'JobRole_Manufacturing Director': np.random.randint(0, 2, 500),
    'JobRole_Research Director': np.random.randint(0, 2, 500),
    'JobRole_Research Scientist': np.random.randint(0, 2, 500),
    'JobRole_Sales Executive': np.random.randint(0, 2, 500),
    'JobRole_Sales Representative': np.random.randint(0, 2, 500),
    'MaritalStatus_Married': np.random.randint(0, 2, 500),
    'MaritalStatus_Single': np.random.randint(0, 2, 500),
    # Add a simple 'JobRole' and 'Department' column back for direct plotting convenience
    'JobRole': np.random.choice(['Sales Executive', 'Research Scientist', 'Laboratory Technician', 'Manager', 'Human Resources'], 500),
    'Department': np.random.choice(['Sales', 'Research & Development', 'Human Resources'], 500)
}
df = pd.DataFrame(data)

# --- Preprocessing ---
# Label encoding for 'Attrition' (converting 'Yes'/'No' to 1/0)
le = LabelEncoder()
df["Attrition_Encoded"] = le.fit_transform(df["Attrition"])

# Ensure the 'Attrition' column used for groupby/unstack is categorical with 0 and 1
df['Attrition'] = pd.Categorical(df['Attrition_Encoded'], categories=[0, 1])

# Create Age Groups for better categorical analysis in RQ1
# These definitions are placed here to ensure they are always available.
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-24', '25-34', '35-44', '45-54', '55+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


print("DataFrame head with new columns:")
print(df.head())

# Define a color map for Attrition (0: No, 1: Yes) for better legend readability
attrition_color_map = {0: 'blue', 1: 'red'}
attrition_label_map = {0: 'No Attrition', 1: 'Yes Attrition'}

# --- Visualization for Research Question 1 (RQ1) ---
# RQ1: How does OverTime affect employee attrition in different job roles and age groups?

print("\n--- Visualizing RQ1: OverTime, Attrition, Job Role, and Age Group ---")

# Calculate attrition rate for each combination of OverTime, JobRole, and AgeGroup
attrition_rates_rq1 = df.groupby(['OverTime', 'JobRole', 'AgeGroup', 'Attrition']).size().unstack(fill_value=0)

# Ensure '0' and '1' columns exist (they should, due to pd.Categorical)
if 1 in attrition_rates_rq1.columns and 0 in attrition_rates_rq1.columns:
    attrition_rates_rq1['Total'] = attrition_rates_rq1[1] + attrition_rates_rq1[0]
    # Calculate attrition rate based on 'Yes' (1) divided by 'Total'
    attrition_rates_rq1['Attrition_Rate'] = (attrition_rates_rq1[1] / attrition_rates_rq1['Total'] * 100).fillna(0)
    attrition_rates_rq1 = attrition_rates_rq1.reset_index()

    fig_rq1 = px.bar(
        attrition_rates_rq1,
        x='JobRole',
        y='Attrition_Rate',
        color='OverTime', # Coloring by OverTime
        facet_col='AgeGroup', # Faceting by AgeGroup
        barmode='group',
        title='Attrition Rate by Job Role, Age Group, and OverTime Status',
        labels={'Attrition_Rate': 'Attrition Rate (%)', 'JobRole': 'Job Role', 'OverTime': 'OverTime'},
        category_orders={"AgeGroup": labels} # Ensure age groups are ordered correctly
    )

    fig_rq1.update_layout(height=600, width=1200)
    fig_rq1.update_yaxes(range=[0, 100]) # Attrition rate is a percentage
    fig_rq1.show()
else:
    print("Cannot calculate attrition rates for RQ1. Encoded '0' or '1' attrition categories missing after categorical conversion.")


# --- Visualization for Research Question 2 (RQ2) ---
# RQ2: Is there a relationship between monthly income and attrition,
#      and does this relationship change depending on the levels of job satisfaction?

print("\n--- Visualizing RQ2: Monthly Income, Attrition, and Job Satisfaction ---")

fig_rq2 = px.scatter(
    df,
    x="MonthlyIncome",
    y="JobSatisfaction", # Using JobSatisfaction on Y-axis as it's ordinal/numerical
    color="Attrition_Encoded", # Use the encoded column for coloring (0/1)
    color_discrete_map=attrition_color_map, # Apply the custom color map for 0/1
    facet_col="JobRole", # Faceting by JobRole to see patterns within roles
    hover_data=['Age', 'OverTime'],
    title="Monthly Income vs. Job Satisfaction by Attrition and Job Role",
    labels={
        "MonthlyIncome": "Monthly Income",
        "JobSatisfaction": "Job Satisfaction (1=Low, 4=Very High)",
        "Attrition_Encoded": "Attrition" # Update label for encoded column in legend
    }
)
# Update legend group names (0 and 1) to meaningful labels ('No Attrition', 'Yes Attrition')
fig_rq2.for_each_trace(lambda t: t.update(name = attrition_label_map[int(t.name)]))
fig_rq2.update_layout(height=800, width=1400)
fig_rq2.update_xaxes(matches=None) # Allow independent x-axes for better detail in facets
fig_rq2.update_yaxes(matches=None) # Allow independent y-axes for better detail in facets
fig_rq2.show()

# --- Alternative for RQ2 if JobSatisfaction is better treated categorically ---
print("\n--- Alternative Visualizing RQ2: Monthly Income, Attrition, and Job Satisfaction (Box Plot) ---")

fig_rq2_alt = px.box(
    df,
    x="JobSatisfaction",
    y="MonthlyIncome",
    color="Attrition_Encoded", # Use the encoded column for coloring (0/1)
    color_discrete_map=attrition_color_map, # Apply the custom color map for 0/1
    facet_col="JobRole",
    title="Monthly Income Distribution by Job Satisfaction, Attrition, and Job Role",
    labels={
        "MonthlyIncome": "Monthly Income",
        "JobSatisfaction": "Job Satisfaction (1=Low, 4=Very High)",
        "Attrition_Encoded": "Attrition" # Update label for encoded column in legend
    },
    category_orders={"JobSatisfaction": [1, 2, 3, 4]} # Ensure order if JobSatisfaction is treated as category
)
# Update legend group names (0 and 1) to meaningful labels ('No Attrition', 'Yes Attrition')
fig_rq2_alt.for_each_trace(lambda t: t.update(name = attrition_label_map[int(t.name)]))
fig_rq2_alt.update_layout(height=800, width=1400)
fig_rq2_alt.update_xaxes(matches=None)
fig_rq2_alt.update_yaxes(matches=None)
fig_rq2_alt.show()
```
